```python
(14)
#Count occurrences of words in a given sentence using dict
sentence  = "Hello india, Hello world , welocme to my country india."
sentence = sentence.lower()
cleaned_sentence = ""
for ch in sentence:
  if ch.isalnum() or ch.isspace():
    cleaned_sentence += ch

words = cleaned_sentence.split()
word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

print(word_counts)
```

```python
#18️⃣ Grade Calculato
marks = int(input("Enter the Number"))

match marks:
  case m if m>=90:
    print("A grade")
  case m if m >= 80:
    print("B Grade")
  case m if m >= 70:
    print("C Grade")
  case m if m >= 60:
    print("D Grade")
  case m if m >= 35:
    print("Pass ")
  case _:
    print("Fail")
print(marks)
```

```python
#20️⃣ Number Guessing Game
import random

num1 = random.randint(1,100)
guess = None
attempt = 0

print("welcome to the guessing the game")
print("I have selected the nummber between 1 to 1000. can you guess it")


while guess != num1:
  try:
    guess = int(input("Enter your guess:"))

    attempt += 1

    if guess < num1:
      print("too low! Try again.")
    elif guess > num1:
      print("Too Hight! Try again.")
    else:
      print(f"Congratualations You gussed the number {num1}, in {attempt} attempts. ")
  except ValueError:
    print("Invalid input. Please enter an integer.")
```
